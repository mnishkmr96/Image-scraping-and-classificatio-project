```python
# importing required libraries
import selenium
import pandas as pd
from selenium import webdriver
import time
import requests
import shutil
```

```python
driver = webdriver.Chrome('C:/Users/user/chromedriver.exe')
```

```python
url = 'https://www.amazon.in/'
driver.get(url)
time.sleep(3)

search_bar = driver.find_element_by_id('twotabsearchtextbox')
search_bar.clear()
search_bar.send_keys('sarees')

search = driver.find_element_by_id('nav-search-submit-button')
search.click()
time.sleep(3)
```

```python
src = []
nxt = 0
while nxt < 4:
    img_tag = driver.find_elements_by_xpath('//div[@class="a-section aok-relative s-image-tall-aspect"]//img[@class="s-image"]')
    for i in img_tag:
        src.append(i.get_attribute('src'))
    try:
        nxt_tag = driver.find_element_by_xpath('//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
        nxt_tag.click()
        time.sleep(3)
    except:
        break
    nxt += 1
print (len (src))
```

```python
a = 1
for image_url in sarees:
    filename = 'sarees_' + str(a) + '.jpg'

    r = requests.get(image_url, stream = True)

# Check if the image was retrieved successfully
    if r.status_code == 200:
    # Set decode_content value to True, otherwise the downloaded image file's size will be zero.
        r.raw.decode_content = True
    
    # Open a local file with wb ( write binary ) permission.
        with open(filename,'wb') as f:
            shutil.copyfileobj(r.raw, f)
        
        print('Image sucessfully Downloaded: ',filename)
    else:
        print('Image Couldn\'t be retreived')
        
    a += 1
```

```python
import os
import glob
```

```python
file_path = 'C:/Users/user/Clothes Classifier/Untitled Folder'
file_type = ['jeans', 'sarees', 'trousers']
file_list = glob.glob(file_path+'/*.*')
print (file_list)
```

```python
# Check if directory exists. If not, create it
for folder in file_type:
    if os.path.isdir(f"./{folder}"):
        pass
    else:
        os.mkdir(f"./{folder}")

# Move files
jeans_folder = 'C:/Users/user/Clothes Classifier/Untitled Folder/jeans'
sarees_folder = 'C:/Users/user/Clothes Classifier/Untitled Folder/sarees'
trousers_folder = 'C:/Users/user/Clothes Classifier/Untitled Folder/trousers'
for file in file_list:
    if file.startswith('C:/Users/user/Clothes Classifier/Untitled Folder\\jeans'):
        shutil.move(file, jeans_folder)
    elif file.startswith('C:/Users/user/Clothes Classifier/Untitled Folder\\sarees'):
        shutil.move(file, sarees_folder)
    elif file.startswith('C:/Users/user/Clothes Classifier/Untitled Folder\\trouser'):
        shutil.move(file, trousers_folder)
```
